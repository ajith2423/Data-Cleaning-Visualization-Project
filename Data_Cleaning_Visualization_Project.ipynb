```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Data-cleaning-project/netflix_titles.csv')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows")

plt.show()
```

```python
plt.figure(figsize=(12,6))

df['release_year'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Release Years")

plt.xlabel("Year")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)

plt.title("Content Ratings Distribution")

plt.show()
```

Insights:
1. Movies are more than TV Shows on Netflix.
2. Most content was released after 2015.
3. TV-MA rating content is highest.
4. Netflix contains content from multiple countries.
